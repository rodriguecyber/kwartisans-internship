## 🐍 Python Data Analysis Lab: BMI & Distributions

**Dataset**: `day06/people_weights.csv`

**Objectives**
- Load and explore data using pandas
- Perform basic data visualization
- Engineer new features (Height and BMI)
- Analyze distributions based on gender


```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# If you're running in Jupyter, keep plots inline.
# (In a .py script, this line is unnecessary.)
%matplotlib inline
```

### Step 1: Load the Data

Load `people_weights.csv` into a DataFrame called `df`.

```python
# Prefer repo-root execution: path includes day06/
path_candidates = [
    "day06/people_weights.csv",
    "people_weights.csv",
]

for p in path_candidates:
    try:
        df = pd.read_csv(p)
        csv_path = p
        break
    except FileNotFoundError:
        df = None

if df is None:
    raise FileNotFoundError(f"Could not find people_weights.csv in any of: {path_candidates}")

print(f"Loaded {len(df)} rows from {csv_path}")
df.head()
```

### Step 2: Basic Visualization

Create a bar chart showing `Name` on the x-axis and `Weight_kg` on the y-axis.

```python
ax = df.plot(kind="bar", x="Name", y="Weight_kg", legend=False, figsize=(10, 4))
ax.set_title("Weight Distribution by Name")
ax.set_ylabel("Weight (kg)")
plt.tight_layout()
plt.show()
```

### Step 3: Feature Engineering (Heights)

Generate random heights between 150cm and 200cm and store them in a new column `Height_cm`.

```python
np.random.seed(42)

df["Height_cm"] = np.random.randint(150, 201, size=len(df))
df.head()
```

### Step 4: Calculate BMI

\[ BMI = \frac{weight\_{(kg)}}{height\_{(m)}^2} \]

Convert height from cm to meters first, then compute BMI into a new `BMI` column.

```python
df["Height_m"] = df["Height_cm"] / 100.0

df["BMI"] = df["Weight_kg"] / (df["Height_m"] ** 2)

df.head()
```

### Step 5: Comparative Analysis

Compare the BMI distribution between Males and Females using histograms (with KDE-style density via `density=True`).

```python
female_bmi = df.loc[df["Sex"].str.lower() == "female", "BMI"]
male_bmi = df.loc[df["Sex"].str.lower() == "male", "BMI"]

plt.figure(figsize=(8, 4))

# Use the same binning for comparability
bins = np.linspace(df["BMI"].min(), df["BMI"].max(), 10)

plt.hist(female_bmi, bins=bins, alpha=0.6, density=True, label="Female")
plt.hist(male_bmi, bins=bins, alpha=0.6, density=True, label="Male")

plt.title("BMI Distribution by Sex")
plt.xlabel("BMI")
plt.legend()
plt.tight_layout()
plt.show()
```

### Quick sanity checks

A small summary to confirm BMI is computed and see basic stats.

```python
df[["Name", "Sex", "Weight_kg", "Height_cm", "BMI"]].sort_values("BMI", ascending=False)

print("\nBMI summary (overall):")
display(df["BMI"].describe())

print("\nBMI summary by sex:")
display(df.groupby("Sex")["BMI"].describe())
```
